```scala
class test extends Module {
    val io = IO(new Bundle {
        val in = Input(Bool())
        val out = Output(UInt(32.W))
    })
    
    io.out := withClock(io.in.asClock) {
        val ctr = Reg(UInt(32.W))
        ctr := ctr + 1.U
        ctr
    }
}

println(getVerilog(new test))
visualize(() => new test)
```
